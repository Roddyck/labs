```python
from main import *
import matplotlib.pyplot as plt
from typing import Callable
from math import sin, cos, cbrt, pi, acos
```

```python
def show_plots(f: Callable, fi: Callable, f_prime: Callable, f_prime_second: Callable,
              a: float, b: float, func_label: str, zero_of_func=None):
    bis_res, points_bis = bisection(f, a, b)
    print(round(bis_res, 3), ' - Метод половинного деления')
    sec_res, points_sec = secant(f, a, b)
    print(round(sec_res, 3), ' - Метод хорд')
    fix_point_res, fix_point_itrs = fixed_point_iteration(fi, f, f_prime_second, a, b)
    print(round(fix_point_res, 3), ' - Метод простой итерации')
    newton_res, newton_itrs = newton(f, f_prime, f_prime_second, a, b)
    print(round(newton_res, 3), ' - Метод Ньютона')

    func_zero = newton_res

    if zero_of_func is not None:
        func_zero = zero_of_func
    
    plt.figure(1, figsize=(9,9))
    plt.subplot(221)
    plt.plot([x / 10.0 for x in range(int(a*10), int(b*10))], [f(x / 10.0) for x in range(int(a*10), int(b*10))], 'b-', label=func_label)
    plt.plot(points_bis['x'], points_bis['y'], 'ro', label='точки итерации')
    plt.grid(True)
    plt.title('Метод половинного деления')
    plt.legend(loc='upper left')
    
    plt.subplot(222)
    plt.plot([x / 10.0 for x in range(int(a*10), int(b*10))], [f(x / 10.0) for x in range(int(a*10), int(b*10))], 'b-', label=func_label)
    plt.plot(points_sec['x'], points_sec['y'], 'ro', label='точки итерации')
    plt.grid(True)
    plt.title('Метод хорд')
    plt.legend(loc='upper left')
    
    plt.subplot(223)
    plt.plot([x / 10.0 for x in range(int(a*10), int(b*10))], [f(x / 10.0) for x in range(int(a*10), int(b*10))], 'b-', label=func_label)
    plt.plot(fix_point_itrs['x'], fix_point_itrs['y'], 'ro', label='точки итерации')
    plt.grid(True)
    plt.title('Метод простой итерации')
    plt.legend(loc='upper left')
    
    plt.subplot(224)
    plt.plot([x / 10.0 for x in range(int(a*10), int(b*10))], [f(x / 10.0) for x in range(int(a*10), int(b*10))], 'b-', label=func_label)
    plt.plot(newton_itrs['x'], newton_itrs['y'], 'ro', label='точки итерации')
    plt.grid(True)
    plt.title('Метод Ньютона')
    plt.legend(loc='upper left')

    plt.figure(2, figsize=(9,9))
    plt.plot([i for i in range(len(points_bis['x']))], [abs(x - func_zero) for x in points_bis['x']], 'b-', label='Метод половинного деления')
    plt.plot([i for i in range(len(points_sec['x']))], [abs(x - func_zero) for x in points_sec['x']], 'g-', label='Метод хорд')
    plt.plot([i for i in range(len(fix_point_itrs['x']))], [abs(x - func_zero) for x in fix_point_itrs['x']], 'y-', label='Метод простой итерации')
    plt.plot([i for i in range(len(newton_itrs['x']))], [abs(x - func_zero) for x in newton_itrs['x']], 'r-', label='Метод Ньютона')
    plt.xlabel('номер итерации')
    plt.ylabel('абсолютная погрешность')
    plt.legend(loc='upper right')
    plt.show()
```

# Функия 1: $ x^3 - x $

Для МПИ нужно привести уравнение к виду $ x = \phi(x)$

Достаточное условие:
$$
    |\phi'(x)| \leq q < 1
$$

Пусть $ \phi(x) = \sqrt[3]{x} $

$$ 
\phi' = \frac{1}{3} x^{-\frac{2}{3}}
$$

$$
x^2 > 27
$$

$$
\left(-\infty, -\frac{\sqrt{3}}{9} \right) \cup \left(\frac{\sqrt{3}}{9}, +\infty \right)
$$

Проверка достаточного условия сходимости метода Ньютона
| [a,b] | [0.5, 2] | +   |
| :-----: | :--------: | :---: |
| $f(a)f(b) < 0$ | $(0.5^3 - 0.5) \cdot (2^3 - 2) < 0$ | + |
| $f'(x) \neq 0$ signf(x) = const | $f'(x) = 3x^2 - 1 = 0$, $ x = \pm \sqrt{1/3} $, sign f(x) $ > 0$ | + | 
| sign $f''(x) = const$ | $f''(x) = 6x > 0$ | + |

Для корня $ x = 0 $ метод Ньютона не будет сходиться на любом отрезке, т.к. $ f''(x) = 6x $ меняет знак в точке $ x = 0 $

```python
def f1(x: float) -> float:
    return x**3 - x

def fi1(x: float) -> float:
    return cbrt(x)

def fi1_2(x: float) -> float:
    return x ** 3

def f1_prime(x: float) -> float:
    return 3 * x ** 2 - 1

def f1_prime_second(x: float) -> float:
    return 6 * x

ZERO_F1_1 = 1.0

show_plots(f1, fi1, f1_prime, f1_prime_second, 0.5, 2, r'$x^3 - x$', ZERO_F1_1)
```

# Функция 2: $x^3 + 3x^2 + 6x - 1$

| x              | $ -\infty $ |  -1 |  1   |  $+ \infty$ |
| ------         | ----------- | --- | ---  | ----------  |
| $ sign f(x) $  |       -     |  -  |  +   |   +         |

$$
x = \frac{-x^3 - 3x^2 + 1}{6} = \phi(x)
$$

Достаточное условие МПИ
$$
\phi'(x) = \frac{1}{6} ( -3x^2 - 6x) = -\frac{1}{2}x^2 - x
$$

Проверка достаточного условия сходимости метода Ньютона
| [a,b] | [-0.5, 0.5] | +   |
| :-----: | :--------: | :---: |
| $f(a)f(b) < 0$ | $(-0.125 + 3 \cdot 0.25 - 3 - 1) \cdot (0.125 + 3 \cdot 0.25 + 3 - 1)  $ | + |
| $f'(x) \neq 0$ $signf'(x) = const $ | $f'(x) = 3x^2 + 6x + 6 = 0$, $sign f'(x)  > 0 $ | + | 
| sign $f''(x) = const$ | $f''(x) = 6x + 6 > 0$ | + |

```python
def f2(x: float) -> float:
    return x**3 + 3 * x**2 + 6 * x - 1
    
def fi2(x: float) -> float:
    return (-x**3 - 3 * x**2 + 1) / 6

def f2_prime(x: float) -> float:
    return 3 * x**2 + 6 ** x + 6

def f2_prime_second(x: float) -> float:
    return 6 * x + 6

show_plots(f2, fi2, f2_prime, f2_prime_second, -0.5, 0.5, r'x^3 + 3x^2 + 6x - 1')
```

# Функция 3: $ x - \sin(x) - 0.25 $

| x        | $-\infty$  |   1   |   2   | $ +\infty$ |
| :---:    | :--------: | :---: | :---: | :-------:  |
| signf(x) |      -     |   -   |   +   |     +      |

$$
x = sin(x) + 0.25 = \phi(x)
$$

$$ \phi'(x) = \cos(x) < 1 \quad \forall x \in [1, 2] $$

Проверка достаточного условия сходимости метода Ньютона
| [a,b] | [1, 2] | +   |
| :-----: | :--------: | :---: |
| $f(a)f(b) < 0$ | $(1 - \sin(1) - 0.25) \cdot (2 - \sin(2) - 0.25)  $ | + |
| $f'(x) \neq 0$ $signf'(x) = const $ | $f'(x) = 1 - \cos(x) \neq 0$, $sign f'(x)  > 0 $ | + | 
| sign $f''(x) = const$ | $f''(x) = sin(x) > 0$ | + |

```python
def f3(x: float) -> float:
    return x - sin(x) - 0.25

def fi3(x: float) -> float:
    return sin(x) + 0.25

def f3_prime(x: float) -> float:
    return 1 - cos(x)

def f3_prime_second(x: float) -> float:
    return sin(x)

show_plots(f3, fi3, f3_prime, f3_prime_second, 1, 2, r'x - sin(x) - 0.25')
```

# Функция 4: $ x - \sin \frac{\pi}{2} x - 0.25 $

| x        | $-\infty$  |   1   |   2   | $ +\infty$ |
| :---:    | :--------: | :---: | :---: | :-------:  |
| signf(x) |      -     |   -   |   +   |     +      |

$$
x = sin\frac{\pi}{2}x + 0.25 = \phi(x)
$$

$$ \phi'(x) = \frac{\pi}{2} \cos \frac{\pi}{2} x < 1 \quad \forall x \in [1, 1.4] $$

Проверка достаточного условия сходимости метода Ньютона
| [a,b]   | [1.1, 1.4]   | +     |
| :-----: | :--------: | :---: |
| $f(a)f(b) < 0$ | $(1.1 - \sin(\frac{\pi}{2} \cdot 1.1) - 0.25) \cdot (1.4 - \sin(\frac{\pi}{2} \cdot 1.4) - 0.25)  $ | + |
| $f'(x) \neq 0$ $signf'(x) = const $ | $f'(x) = 1 - \frac{\pi}{2}\cos(\frac{\pi}{2}x) = 0$, $sign f'(x) < 0 $ | + | 
| sign $f''(x) = const$ | $f''(x) = \frac{\pi^2}{4}\sin(\frac{\pi}{2}x) > 0$ | + |

```python
def f4(x: float) -> float:
    return x - sin(pi / 2 * x) - 0.25

def fi4(x: float) -> float:
    return sin(pi / 2 * x) + 0.25

def f4_prime(x: float) -> float:
    return 1 - pi / 2 * cos(pi / 2 * x)

def f4_prime_second(x: float) -> float:
    return pi**2 / 4 * sin(pi / 2 * x)
    
show_plots(f4, fi4, f4_prime, f4_prime_second, 1.1, 1.4, r'x - sin(pi/2 * x) - 0.25')
```

# Функия 5: $ x - \cos x $

| x        | $-\infty$  |   0.5 |   1   | $ +\infty$ |
| :---:    | :--------: | :---: | :---: | :-------:  |
| signf(x) |      -     |   -   |   +   |     +      |

$$
x = \cos x = \phi(x)
$$

$$ \phi'(x) = -\sin x < 1 \quad \forall x \in [0.5, 1] $$

Проверка достаточного условия сходимости метода Ньютона
| [a,b]   | [0.5, 1]   | +     |
| :-----: | :--------: | :---: |
| $f(a)f(b) < 0$ | $(0.5 - \cos(0.5)) \cdot (1 - \cos (1))  $ | + |
| $f'(x) \neq 0$ $signf'(x) = const $ | $f'(x) = 1 + \sin(x) \neq 0$, $sign f'(x) > 0 $ | + | 
| sign $f''(x) = const$ | $f''(x) = \cos(x) > 0$ | + |

```python
def f5(x: float) -> float:
    return x - cos(x)

def fi5(x: float) -> float:
    return cos(x)

def f5_prime(x: float) -> float:
    return 1 + sin(x)

def f5_prime_second(x: float) -> float:
    return cos(x)
    
show_plots(f5, fi5, f5_prime, f5_prime_second, 0.5, 1.0, r'x - cos(x)')
```

# Функция 6: $ x - \cos \frac{\pi}{2} x $

| x        | $-\infty$  |   0.5 |   1   | $ +\infty$ |
| :---:    | :--------: | :---: | :---: | :-------:  |
| signf(x) |      -     |   -   |   +   |     +      |

$$
cos \frac{\pi}{2} x = x
$$

$$ x = \frac{2\arccos x}{\pi} = \phi(x) $$

$$ \phi'(x) = -\frac{2}{\pi \sqrt{1-x^2}} < 1 \quad \forall x \in [0.5, 0.7] $$

Проверка достаточного условия сходимости метода Ньютона
| [a,b]   | [0.5, 0.7]   | +     |
| :-----: | :--------: | :---: |
| $f(a)f(b) < 0$ | $(0.5 - \cos(\frac{\pi}{2} \cdot 0.5)) \cdot (0.7 - \cos (\frac{\pi}{2} \cdot 0.7))  $ | + |
| $f'(x) \neq 0$ $signf'(x) = const $ | $f'(x) = 1 + \frac{\pi}{2}\sin(\frac{\pi}{2}x) \neq 0$, $sign f'(x) > 0 $ | + | 
| sign $f''(x) = const$ | $f''(x) = \frac{\pi^2}{4}\cos(\frac{\pi}{2}x) > 0$ | + |

```python
def f6(x: float) -> float:
    return x - cos(pi / 2 * x)

def fi6(x: float) -> float:
    return 2 * acos(x) / pi

def f6_prime(x: float) -> float:
    return 1 + pi / 2 * sin(pi / 2 * x)

def f6_prime_second(x: float) -> float:
    return pi**2 / 4 * cos(pi / 2 * x)
    
show_plots(f6, fi6, f6_prime, f6_prime_second, 0.5, 0.7, r'x - cos(pi/2 * x)')
```
